```python
import numpy as np
```

```python
N = 100 # Number of samples
D = 2 # Number of features
```

```python
X = np.random.randn(N,D)
X
```

```python
# Subtract 2 from the first 50 rows
X[:50, :] = X[:50, :] - 2*np.ones((50,D))
```

```python
# Add 2 to the last 50 rows
X[50:, :] = X[50:, :] + 2*np.ones((50,D))
```

```python
X
```

```python
# Create array of targets
T = np.array([0]*50 + [1]*50)
```

```python
# Concatenate a column of ones
ones = np.array([[1]*N]).T
Xb = np.concatenate((ones, X), axis=1)
```

```python
# randomly initialize the weights
w = np.random.randn(D+1)
w
```

```python
# calculate the model output
z = Xb.dot(w)
z
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
Y = sigmoid(z)
Y
```

```python
def cross_entropy(T, Y):
    E = 0
    for i in range(N):
        if T[i] == 1:
            E -= np.log(Y[i])
        else:
            E -= np.log(1 - Y[i])
    return E
```

```python
print(cross_entropy(T, Y))
```
