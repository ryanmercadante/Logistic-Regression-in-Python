```python
import numpy as np
```

```python
N = 100
D = 2
```

```python
X = np.random.randn(N,D)
X
```

```python
ones = np.array([[1]*N]).T
Xb = np.concatenate((ones, X), axis=1)
Xb
```

```python
# randomly initialize a weight vector
w = np.random.randn(D+1)
w
```

```python
# calculate the dot product between each row of x and w
z = Xb.dot(w)
z.shape
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
print(sigmoid(z))
```
